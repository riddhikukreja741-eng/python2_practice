```python
#LISTS

list1=['a','b','c','d']
list2=['apple','ball','dog','cat']

#APPEND
list1.append('e')
print(list1)
list1.append([1,2,3])
print(list1)
list1.append(7)
print(list1)
list1.append(True)
print(list1)
list1.append(2.0)
print(list1)

#EXTEND
list1.extend(['ABC','DEF'])
print(list1)

#INSERT
list1.insert(1,'XYZ')
print(list1)

list1.insert(20,'XYZ')
print(list1)

#POP
list1.pop()
print(list1)

list1.pop(1)
print(list1)

#REMOVE
list1.remove('ABC')
print(list1)

#SORT
list1.sort
print(list1)                       #ascending order

list2.sort
print(list2)

list2.sort(reverse=True)             #descending order
print(list2)

list1.reverse()
print(list1)

for i in list1:
  print(type(i))

list=['a','b','c','d']
print(list[2])
```

```python
#DICTIONARY AND SETS

list1=[1,1,2,3,4,3]
s=set(list1)
print(s)
print(type(s))

t=tuple(list1)
print(t)
print(type(t))

s1=set('aabbccddee')
print(s1)

s2={0,1,'1',True,'True'}
print(s2)

D={"Name":"Riddhi",
   "Rollno":10,
   "address":"Mumbai"}

print(D)
print(D["Rollno"])

listA=[1,2,3]
listB=['a','b','c']

D1=dict(zip(listA,listB))
print(D1)

for x in D.keys():
    print(x)

for y in D.values():
    print(y)

for x,y in D.items():
    print(x,y)
```
