```python
# Question 5: Unique ID Management
# You have two lists of employee IDs:
# dept_a = ["E01", "E02", "E03", "E01"]
# dept_b = ["E03", "E04", "E05", "E02"]
# 1. Convert both lists into sets to remove duplicates.
# 2. Find all unique IDs that exist in both departments (Intersection).
# 3. Find all unique IDs that exist across all departments (Union).
# 4. Find the IDs that are only in dept_a but not in dept_b (Difference)

dept_a = ["E01", "E02", "E03", "E01"]
dept_b = ["E03", "E04", "E05", "E02"]

set_a=set(dept_a)
set_b=set(dept_b)
intersection=set_a.intersection(set_b)
union=set_a.union(set_b)
difference=set_a.difference(set_b)
print(intersection)
print(union)
print(difference)
```
